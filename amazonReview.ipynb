```
import pandas as pd
import gzip
```

```
data = pd.read_json('http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz', lines =  True)
```

```
data.head(10)
```

```
dF = pd.DataFrame(data)
```

```
def wordCount(str):
  count = dict()
  words = str.split()
  for w in words:
    if w in count:
      count[w]+=1
    else:
      count[w]=1
  return count
```

```
dF['reviewText_wordCount'] = dF['reviewText'].apply(wordCount)
```

```
dF['reviewText_wordCount'] = dF['reviewText'].apply(wordCount)
```

```
dF['summary_wordCount'] = dF['summary'].apply(wordCount)
```
